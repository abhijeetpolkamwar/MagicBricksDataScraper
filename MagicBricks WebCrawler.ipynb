```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

url = input('Enter the MagicBricks URL')
#url = 'https://www.magicbricks.com/property-for-sale/residential-real-estate?bedroom=&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment&cityName=West-area-Pune'
results = requests.get(url)
soup = BeautifulSoup(results.text,'lxml')

#SOCIETY AND PRICE COLUMN DATA
price = []
society = []
p = soup.find_all('div',class_='mb-srp__card__price--amount')
for i in p:
    a = i.getText()
    price.append(a)

s = soup.find_all('div',class_='mb-srp__card__society')
for i in s:
    b = i.getText()
    society.append(b)
    

#CARPET AREA COLUMN DATA

carpet_area_raw = []
car = []
carpet_area = []
sup = []
super_area_raw = []
super_area = []

for i in soup.find_all('div',{'data-summary':'carpet-area'}) or  soup.find_all('div',{'data-summary':'super-area'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    car.append(z)
    
for i in car:
    a = i.getText()
    carpet_area_raw.append(a)
    
for i in carpet_area_raw:
    n = i.split()[0]
    carpet_area.append(n)

"""
if soup.find_all('div',{'data-summary':'carpet-area'}):
    for i in soup.find_all('div',{'data-summary':'carpet-area'}):
        z = i.select('div',class_="mb-srp__card__summary--value")[1]
        car.append(z)
    
    for i in car:
        a = i.getText()
        carpet_area_raw.append(a)
    
    for i in carpet_area_raw:
        n = i.split()[0]
        carpet_area.append(n)
else:
    for i in soup.find_all('div',{'data-summary':'super-area'}):
        z = i.select('div',class_="mb-srp__card__summary--value")[1]
        sup.append(z)
    
    for i in car:
        a = i.getText()
        super_area_raw.append(a)
    
    for i in carpet_area_raw:
        n = i.split()[0]
        super_area.append(n)
"""    
#BATHROOM COLUMN DATA
bathroom_raw = []
bath = []
bathroom = []

for i in soup.find_all('div',{'data-summary':'bathroom'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    bath.append(z)
    
for i in bath:
    a = i.getText()
    bathroom_raw.append(a)
    
for i in bathroom_raw:
    n = i.split()[0]
    bathroom.append(n)


#STATUS COLUMN DATA

status_raw = []
stat = []
status = []

for i in soup.find_all('div',{'data-summary':'status'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    status.append(z.getText())
    
#for i in stat:
#    a = i.getText()
#    status_raw.append(a)
    
#for i in status_raw:
#    n = i.split()[0]
#    status.append(n)


#TRANSACTION COLUMN DATA

transaction_raw = []
trans = []
transaction = []

for i in soup.find_all('div',{'data-summary':'transaction'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    trans.append(z)
    
for i in trans:
    a = i.getText()
    transaction_raw.append(a)
    
for i in transaction_raw:
    n = i.split()[0]
    transaction.append(n)
    

#FURNISHING COLUMN DATA

furnishing_raw = []
furnish = []
furnishing = []

for i in soup.find_all('div',{'data-summary':'furnishing'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    furnish.append(z)
    
for i in furnish:
    a = i.getText()
    furnishing_raw.append(a)
    
for i in furnishing_raw:
    n = i.split()[0]
    furnishing.append(n)
    
    
#FACING COLUMN DATA

facing_raw = []
face = []
facing = []

for i in soup.find_all('div',{'data-summary':'facing'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    face.append(z)
    
for i in face:
    a = i.getText()
    facing_raw.append(a)
    
for i in facing_raw:
    n = i.split()[0]
    facing.append(n)
    
    
#OVERLOOKING COLUMN DATA

overlooking_raw = []
overlook = []
overlooking = []

for i in soup.find_all('div',{'data-summary':'overlooking'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    overlook.append(z)
    
for i in overlook:
    a = i.getText()
    overlooking_raw.append(a)
    
for i in overlooking_raw:
    n = i.split()[0]
    overlooking.append(n)
    
    
#OWNERSHIP COLUMN DATA

ownership_raw = []
owner = []
ownership = []

for i in soup.find_all('div',{'data-summary':'ownership'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    owner.append(z)
    
for i in owner:
    a = i.getText()
    ownership_raw.append(a)
    
for i in ownership_raw:
    n = i.split()[0]
    ownership.append(n)
    
    
#PARKING COLUMN DATA

parking_raw = []
park = []
parking = []

for i in soup.find_all('div',{'data-summary':'parking'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    park.append(z)
    
for i in park:
    a = i.getText()
    parking_raw.append(a)
    
for i in parking_raw:
    n = i.split()[0]
    parking.append(n)
    
    
#USP COLUMN DATA

usp_raw = []
up = []
usp = []

for i in soup.find_all('div',{'data-summary':'usp'}):
    z = i.select('div',class_="mb-srp__card__summary--value")[1]
    up.append(z)
    
for i in up:
    a = i.getText()
    usp_raw.append(a)
    
for i in usp_raw:
    n = i.split()[0]
    usp.append(n)
    
    
#DESCPRIPTION COLUMN DATA

desc = []
description = []

for i in soup.find_all('div',{'class':'mb-srp__card--desc--text'}):
    a=i.find_all('p')
    desc.append(a)
    
for i in desc:
    a = i[0].getText()
    description.append(a)
    
    
#TITLE COLUMN

rum = []
for i in soup.find_all('h2',{'class':'mb-srp__card--title'}):
    a = i.getText()
    rum.append(a)
    
#ROOMS COLUMS
number_of_rooms = []
for i in rum:
    a = i[0][0]
    number_of_rooms.append(a)

    
#CITY COLUMN
city = []

for i in rum:
    a = i.split()[-1]
    city.append(a)
    
    
#LOCALITY COLUMN
locality = []

for i in rum:
    a = i.split('in ')[1].split(',')[1:]
    str1 = ','
    b = str1.join(a)
    locality.append(b)
    
    
Society = pd.Series(society)
City = pd.Series(city)
Locality = pd.Series(locality)
Rooms = pd.Series(number_of_rooms)
Price = pd.Series(price)
Carpet_Area = pd.Series(carpet_area)
Bathroom = pd.Series(bathroom)
Status = pd.Series(status)
#Floor = pd.Series(floor)
Transaction = pd.Series(transaction)
Furnishing = pd.Series(furnishing)
Facing = pd.Series(facing)
Overlooking = pd.Series(overlooking)
Ownership = pd.Series(ownership)
Parking = pd.Series(parking)
#Balcony = pd.Series(balcony)
USP = pd.Series(usp)
Description = pd.Series(description)


frame = {'Society':Society,'Locality':Locality,'City':City,'Carpet_Area(sqft)':Carpet_Area,'Rooms':Rooms,'Bathroom':Bathroom,
         'Parking':Parking,'Status':Status, 'Transaction':Transaction,'Furnishing':Furnishing,'Price':Price }

df = pd.DataFrame(frame)
```
